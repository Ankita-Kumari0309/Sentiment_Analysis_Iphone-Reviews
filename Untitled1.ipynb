## Sentiment Analysis on Iphone reviews

## Data Preprocessing 

Before moving on to sentiment analysis, it's important to clean and prepare the text data properly. So, I started with some basic preprocessing steps to improve the quality of the dataset. First, I checked for any null or duplicate values and handled them to ensure clean input. Then, I converted all the text to lowercase for consistency. After that, I removed stopwords, punctuation, and special characters to eliminate noise from the data.

Once the text was cleaned, I tokenized it to break it into individual words. Finally, I applied lemmatization to convert each word to its root form, which helps in better understanding and analysis of the reviews.

#### Remove the stopwords 

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
def removing_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in stop_words])
data['review_text']=data['review_text'].apply(lambda text:removing_stopwords(text))
data['review_text'].head()
```

#### Removing punctuation marks , special characters

```python
import string
punctuators=string.punctuation
punctuations_list=punctuators
def remove_punctuations(text):
    translated=str.maketrans('', '', punctuations_list)
    return text.translate(translated)
data['review_text']= data['review_text'].apply(lambda x: remove_punctuations(x))
data['review_text'].head()
```

# Applying Lemmatization
Lemmatization means converting words to their root or dictionary form, like changing "running" to "run." It helps treat different forms of the same word as one, making the analysis easier and more accurate. I used lemmatization instead of stemming because stemming just cuts off word endings and can create non-words, while lemmatization gives real base words. Since TextBlob works better with proper base words for sentiment analysis, lemmatization was a better choice.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]

data['tokens'] = data['tokens'].apply(lemmatize_tokens)
```

```python
data.head()
```

# Subjectivity and Polarity
Polarity measures if a text is positive, negative, or neutral (from -1 to 1).
Subjectivity shows if a text is opinion-based (1) or fact-based (0).

I used them to understand the sentiment and how personal the reviews are.

```python
def getSubjectivity(text):
    text=' '.join(text)
    return TextBlob(text).sentiment.subjectivity

# polarity
def getpolarity(text):
    text=' '.join(text)
    return TextBlob(text).sentiment.polarity
# creatinf two columns two stope subjecticity and polarity
data['subjectivity']=data['tokens'].apply(getSubjectivity)
data['polarity']=data['tokens'].apply(getpolarity)
```

```python
data.head()
```

# computing positive ,negative and neutral reviews

```python
def getAnalysis(score):
    if score<0:
        return 'negative'
    elif score==0:
        return 'neutral'
    else:
        return 'positive'
        
data['analysis']=data['polarity'].apply(getAnalysis)
data.head()
```

## Vader model
VADER (Valence Aware Dictionary and sEntiment Reasoner) is a rule-based sentiment analysis tool that works well on short, informal text like social media posts.

```python
# Import necessary library
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
vader_analyzer = SentimentIntensityAnalyzer()

# Define a function to get sentiment scores using VADER
def get_vader_sentiment(text):
    scores = vader_analyzer.polarity_scores(str(text))
    return pd.Series([scores['pos'], scores['neu'], scores['neg'], scores['compound']])

# Apply the function to the 'review_text' column
data[['vader_pos', 'vader_neu', 'vader_neg', 'vader_compound']] = data['review_text'].apply(get_vader_sentiment)

# Create sentiment label based on compound score
def vader_sentiment_label(compound_score):
    if compound_score >= 0.05:
        return 'positive'
    elif compound_score <= -0.05:
        return 'negative'
    else:
        return 'neutral'

data['vader_sentiment'] = data['vader_compound'].apply(vader_sentiment_label)

# View sentiment distribution
print(data['vader_sentiment'].value_counts(normalize=True) * 100)
```
